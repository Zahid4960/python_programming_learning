New column addition

```python
# Add new column max correct
def get_max_correct(value):
    print(value)
    return max(value)

bcs_df['Max Correct'] = bcs_df[['A', 'B', 'C', 'D']].idxmax(axis=1)
bcs_df.head()
```

44th BCS data analysis

```python
# subject wise correct answer analysis for 44th bcs
bcs_44_df = bcs_df[bcs_df['Bcs'].isin([44])]
subject_grouped_44_bcs_df = bcs_44_df.groupby(by=['Subject'])[['A', 'B', 'C', 'D']].sum()
subject_grouped_44_bcs_df
```

```python
# bar chart for subject wise correct answer 44th bcs
subject_grouped_44_bcs_df.plot(kind='bar', xlabel='Subject', ylabel='No of correct answer', title='Subject wise correct answer - 44th BCS', figsize=(10, 5))
```
