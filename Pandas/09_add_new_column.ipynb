```python
import pandas as pd

# read csv file
student_performance_df = pd.read_csv('students_performance.csv')
student_performance_df.head()

# calculate language score 
language_score = ((student_performance_df['math score'] + student_performance_df['reading score'] + student_performance_df['writing score']) / 3).astype(int)
language_score

# get max, min & len for language score
max(language_score)
min(language_score)
len(language_score)

# add new column language score
student_performance_df['language score'] = language_score
student_performance_df.head()
```
