```python
import pandas as pd

# read data from csv file
raw_population_df = pd.read_csv('F:\Python_Programming_Learning\Pandas\Data\population_total.csv')
raw_population_df.head()
```

```python
# get shape
raw_population_df.shape
```

```python
# get size
raw_population_df.size
```

```python
# clean data by droping null values
processed_population_df = raw_population_df.dropna()
processed_population_df
```

```python
# get process data's shape
processed_population_df.shape
```

```python
# get process data's size
processed_population_df.shape
```

```python
# make pivot by using proccessed data
population_pivot = processed_population_df.pivot(index='year', columns='country', values='population')
population_pivot
```

```python
# select some countries
selected_countries_pivot_df = population_pivot[ ['Afghanistan', 'Bangladesh', 'India', 'Pakistan', 'Indonesia', 'Iran', 'Brazil'] ]
selected_countries_pivot_df
```

```python
# visualize data with line plot
selected_countries_pivot_df.plot(kind='line', xlabel='year', ylabel='population (Billion)', title='Population data analysis (1960 - 2020)', figsize=(10,5))
```

```python
# select one year data only
selected_countries_pivot_df_for_year_2020 = selected_countries_pivot_df[selected_countries_pivot_df.index.isin([2020])]
selected_countries_pivot_df_for_year_2020
```

```python
# transpose dataframe rows to cols or viseversa
selected_countries_pivot_df_for_year_2020 = selected_countries_pivot_df_for_year_2020.T
selected_countries_pivot_df_for_year_2020
```

```python
# visualize data with bar chart
selected_countries_pivot_df_for_year_2020.plot(kind='bar', xlabel='country', ylabel='population (billion)', title='Population data analysis - 2020', figsize=(10, 5), color='red')
```

```python
# select data only for year 1960, 1980, 2000, 2020 of some countries
selected_years_pivot_df = selected_countries_pivot_df[selected_countries_pivot_df.index.isin([1960, 1980, 2000, 2020])].T
selected_years_pivot_df
```

```python
# visualize data with grouped bar chart
selected_years_pivot_df.plot(kind='bar', xlabel='country', ylabel='population (billion)', title='Population analysis', figsize=(10, 5))
```

```python
# visualize data with pie chart
rename_2020_col_to_str = selected_countries_pivot_df_for_year_2020.rename(columns={2020: '2020'})
rename_2020_col_to_str.plot(kind='pie', y='2020', title='population in 2020 (%)')
```
