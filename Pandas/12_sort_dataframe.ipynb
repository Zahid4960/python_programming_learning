```python
import pandas as pd

# read csv file
student_performance_df = pd.read_csv('students_performance.csv')
student_performance_df.head()

# sort by single column for number
student_performance_df.sort_values(by='math score')
student_performance_df.sort_values(by='math score', ascending=False)

# sort by multiple coulmn for number
student_performance_df.sort_values(by=['math score', 'reading score'], ascending=False)
student_performance_df.head()

# sort by single column for string
student_performance_df.sort_values(by='race/ethnicity', ascending=False, key=lambda col:col.str.lower())
student_performance_df.head()

# sort by multiple column for string
student_performance_df.sort_values(by=['race/ethnicity', 'lunch'], inplace=True, key=lambda col:col.str.lower())
student_performance_df.head()
```
